# How to plot two line in one plot  

```python
import matplotlib.pyplot as plt
x=[2,4,18]
y=[12,34,56]
x2=[3,10,45]
y2=[7,45,30]
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

# How to change plot style 

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')    #any other diffretnt style plz check google
x=[2,4,18]
y=[12,34,56]
x2=[3,10,45]
y2=[7,45,30]
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

# Labels

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[2,4,18]
y=[12,34,56]
x2=[3,10,45]
y2=[7,45,30]
plt.plot(x,y, label='first')
plt.plot(x2,y2,label='second')
plt.legend()   #first or second is a legand
plt.show()
```

# Linewidth

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[2,4,18]
y=[12,34,56]
x2=[3,10,45]
y2=[7,45,30]
plt.plot(x,y, label='first',linewidth=5,c='g')  #change line colour
plt.plot(x2,y2,label='second',linewidth=8)
plt.legend()
plt.show()
```

# Scatter plot

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[2,4,18]
y=[12,34,56]
x2=[3,10,45]
y2=[7,45,30]
plt.scatter(x,y, label='first',linewidth=5)
plt.plot(x2,y2,label='second',linewidth=8)
plt.legend()
plt.show()
```
